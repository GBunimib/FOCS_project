```python
import pandas as pd
import numpy as np
import datetime as dt
from datetime import date 
import random
```

**1. Normalize the loan_lenders table. In the normalized table, each row must have one loan_id and one lender.**

```python
loans_lenders = pd.read_csv("C:/Users/giuli/Google Drive/DS/Foundation of computer science/Progetto/loans_lenders.csv")
loans_lenders.head()
```

```python
#Lista contenente la lista di prestatori che partecipano ad ogni prestito
liste = [loans_lenders.lenders[i].split(',') for i in range(len(loans_lenders))]

#Lista unica contentente tutti i prestatori
lista = [item for elem in liste for item in elem]  
```

```python
#Dataframe contenente i prestatori

Lender = pd.DataFrame(lista, columns = ['permanent_name']) 
Lender.head()
```

```python
#Conteggio di quanti prestatori partecipano ad ogni prestitio (loan_id)
lunghezze = [len(elem) for elem in liste]   

#Replica il numero id del prestito tante volte quanti sono i prestatori che vi partecipano
rep = np.repeat(loans_lenders.loan_id, lunghezze)  
```

```python
#Dataframe che contiene le ripetizioni degli id per numero di lender

rep_id = pd.DataFrame(rep).reset_index() 
rep_id = rep_id.drop('index', 1)
rep_id.head()
```

```python
#Unisce le due liste cos√¨ da ottenere un id del prestito per ogni prestatore che vi partecipa

loan_lender = pd.concat([rep_id, Lender], axis = 1)
loan_lender.head()
```

**2. For each loan, add a column duration corresponding to the number of days between the disburse time and the planned expiration time. If any of those two dates is missing, also the duration must be missing.**

```python
loans = pd.read_csv("C:/Users/giuli/Google Drive/DS/Foundation of computer science/Progetto/loans.csv",
                 parse_dates = ['disburse_time', 'planned_expiration_time'] ) 
loans.head()
```

```python
#Differenza tra le colonne interessate ed estrae il valore giorno, i valori nulli sono sostituiti con 0

loans['duration'] = (loans['planned_expiration_time'] - loans['disburse_time']).apply(lambda x: x.days).fillna(0)
loans.head()
```

**3. Find the lenders that have funded at least twice.**

```python
#Conta quante volte un prestatore ha partecipato ai prestiti

lender = pd.DataFrame(loan_lender.groupby('permanent_name').count()['loan_id']) 
lender = lender.rename(columns = {'loan_id':'count_lender'})
lender.head()
```

```python
#Perstatori che hanno prestato pi√π di una volta

good_lender = lender[lender['count_lender'] > 1] 
good_lender.head()
```

**4. For each country, compute how many loans have involved that country as borrowers**

```python
#Conta il numero di prestiti per stato

country_loans = pd.DataFrame(loans.groupby(['country_name']).count()['loan_id']) 
country_loans = country_loans.rename(columns = {'loan_id':'count_loans'})
country_loans.head()
```

**5. For each country, compute the overall amount of money borrowed.**

```python
#Somma dell'importo prestato ad ogni stato

country_amount = pd.DataFrame(loans.groupby(['country_code']).sum()['funded_amount']) 
country_amount = country_amount.rename(columns = {'funded_amount':'funded_country'})
country_amount.head()
```

**6. Like the previous point, but expressed as a percentage of the overall amount lent**


```python
#Totale importo prestato

total_lent = country_amount.sum(axis = 0)['funded_country'] 
total_lent
```

```python
#Calcola il valore percentuale: totale prestato ad ogni stato / totale complessivo prestato * 100

country_amount['funded_country%'] = country_amount['funded_country'] / total_lent * 100  
country_amount.head()
```

**7. Like the three previous points, but split for each year (with respect to disburse time)**

```python
#Estrae l'anno di pagamento

loans['disburse_year'] = loans.disburse_time.dt.year.astype('Int64')
loans.head()
```

7.1. For each country, compute how many loans have involved that country as borrowers, split for each year

```python
#Conta il numero di prestiti ad ogni stato diviso per anno

state_year_loans = pd.DataFrame(loans.groupby(['country_name', 'disburse_year']).count()['loan_id']) 
state_year_loans = state_year_loans.rename(columns = {'loan_id':'count_loans'})
state_year_loans.head()
```

7.2. For each country and year, compute the overall amount of money borrowed.

```python
#Somma per stato e anno dell'importo preso in prestito

state_year_amount = pd.DataFrame(loans.groupby(['country_name', 'disburse_year']).sum()['funded_amount']) 
state_year_amount = state_year_amount.rename(columns = {'funded_amount':'funded_country_year'})
state_year_amount.head()
```

7.3. Like the previous point, but expressed as a percentage of the overall amount lent in the year

```python
#Totale preso in prestito per anno

total_lent_year = pd.DataFrame(loans.groupby('disburse_year').sum()['funded_amount']) 
total_lent_year = total_lent_year.rename(columns={'funded_amount':'funded_year'})
total_lent_year
```

```python
#Calcola il valore percentuale: importo prestato allo stato nell'anno / totale complessivo per anno * 100

state_year_amount['funded%'] = state_year_amount['funded_country_year'] / total_lent_year['funded_year'] * 100 
state_year_amount.head()
```

**8. For each lender, compute the overall amount of money lent. For each loan that has more than one lender, you must assume that all lenders contributed the same amount.**

**9. For each country, compute the difference between the overall amount of money lent and the overall amount of money borrowed. Since the country of the lender is often unknown, you can assume that the true distribution among the countries is the same as the one computed from the rows where the country is known.**

```python
lenders = pd.read_csv('C:/Users/giuli/Google Drive/DS/Foundation of computer science/Progetto/lenders.csv')
lenders.head()
```

```python
#Lista contenente gli stati non nulli

country = list(lenders['country_code'].dropna())
```

```python
#Creazione di un dizionario in cui, ad ogni stato, √® associata la percentuale con la quale si presenta all'interno di lenders

proporzioni = {}
for stato in country:
  if stato not in proporzioni:
    proporzioni[stato] = country.count(stato) / len(country) * 100
```

```python
#Sostituisce ai valori NA di country_code uno stato (chiave) scelto dal dizionario proporzioni,
#proporzionalmente alla frequenza (valore) con cui si presenta

lenders["country_code"].fillna(random.choices(list(proporzioni.keys()), weights = list(proporzioni.values()), k = 1)[0], inplace = True) 
lenders.head()
```

```python
#Estrae colonne di interesse da lenders per velocizzare il procedimento

small_lenders = lenders[['permanent_name', 'country_code']].copy()
small_lenders.head()
```

```python
#Unisce il dataset creato sopra con lender_lent (conentente il quantitativo prestato da ogni prestatore) su permanent_name

merged1 = pd.merge(small_lenders, lender_lent, on = 'permanent_name')
merged1.head()
```

```python
#Somma per stato di quanto √® stato prestato

tot_lent_country = pd.DataFrame(merged1.groupby('country_code').sum())
tot_lent_country = tot_lent_country.rename(columns={'amount_lender':'lent_country'})
tot_lent_country.head()
```

```python
#Unisce il dataset sopra con country_amount (contentente l'importo prestato ad ogni stato) su country_code

country_diff = pd.merge(tot_lent_country, country_amount, on = 'country_code')
country_diff.head()
```

```python
#Calcola la differenza tra quanto √® stato prestato da ogni stato e quanto √® stato preso in prestito

country_diff['difference'] = country_diff['lent_country'] - country_diff['funded_country']
country_diff.head()
```

**12. For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.**

```python
#Estrae da loans le colonne di interesse per velocizzare il procedimento

ridotto=loans[['loan_id','disburse_time', 'planned_expiration_time', 'loan_amount', 'duration', 'disburse_year']].copy()
ridotto.head()
```

```python
#Estrae il valore anno dalla data di fine prestito e pone a 0 gli anni che sono mancanti

ridotto['disburse_year'] = ridotto['disburse_year'].fillna(0)
ridotto['planned_expiration_year'] = ridotto['planned_expiration_time'].apply(lambda x: x.year).fillna(0).astype('Int64')
ridotto.head()
```

```python
#Creazione di un dizionario che associa ad ogni anno quanto √® stato prestato.

#Nel caso in cui la durata del prestito sia sconosciuta:
#-se l'anno di fine √® sconosciuto allora si associa l'intero ammontare del prestito all'anno di inizio dello stesso;
#-se l'anno di inizio √® sconosciuto allora si associa l'intero ammonatare del prestito all'anno di fine dello stesso.

anni = {}
for row in range(len(ridotto)):
  if ridotto['duration'][row] <= 0 :
    if ridotto['planned_expiration_year'][row]== 0:
        anno = ridotto['disburse_year'][row]
        if anno not in anni:
          anni[anno] = ridotto['loan_amount'][row]
        else:
          anni[anno] += ridotto['loan_amount'][row]
        
    else:
        anno = ridotto['planned_expiration_year'][row]
        if anno not in anni:
          anni[anno] = ridotto['loan_amount'][row]
        else:
          anni[anno] += ridotto['loan_amount'][row]

  else:
    for i in range(int(ridotto['duration'][row]) + 1):
      giorno = ridotto['disburse_time'][row] + dt.timedelta(days = i)
      amount_per_day = ridotto['loan_amount'][row] / ridotto['duration'][row]
      if giorno.year not in anni: 
        anni[giorno.year] = amount_per_day 
      else:
        anni[giorno.year] += amount_per_day 

anni
```

```python
year_loans = pd.DataFrame(anni.items(), columns = ['Year', 'Amount']).sort_values(by = 'Year')
year_loans
```
